# **Modelo Isotrópico y Cuadrado de Ising en 2D con Condiciones de Frontera Periódicas e Interacciones de Vecinos más Cercanos**

## **1. Definición Usual del Lattice**

Consideramos una lattice (red/retículo) cuadrada $Z^2$ y una cuadrado finito $\Lambda \subset Z^2$ centrado alrededor del origen con $|\Lambda| = n^2$ sitios de la red. En particular, note que se pueden rotular los sitios que pertenecen a $\Lambda$ empezando desde una esquina del cuadrado y definiendo filas en una de las direcciones y columnas en la otra (por ejemplo usando la definición de una red que utiliza combinaciones lineales enteras de ciertos vectores base), permitiendo escribir $\Lambda = \{(i, j) | i, j \in \{1, 2, \ldots, n\}\}$. De ahora en adelante el cuadrado es a lo que me referiré como "el sistema".

A cada sitio $x \in \Lambda$ le asociamos una variable de espín $s_x \in \{-1, 1\}$.

Una configuración del sistema es una asignación particular de espines a cada sitio del cuadrado $\Lambda$, es decir, una función $s: \Lambda \to \{-1, 1\}$. Luego, el conjunto de todas las configuraciones posibles es $\Omega_{\Lambda} = \{s: \Lambda \to \{-1, 1\}\}$.

De forma equivalente, podemos pensar en una configuración del sistema como una tupla de $n^2$ elementos, es decir, $(s_{i, j})_{i, j \in \{1, 2, \ldots, n\}}$, donde $s_{i, j} = s(i, j) \in \{-1, 1\}$ para $s \in \Omega_{\Lambda}$; note que en esta representación el conjunto de todas las configuraciones posibles es simplemente {$-1, 1$}$^{n^2}$.

Si ahora agrupamos las configuraciones por filas, entonces podemos definir la configuración $\mu_{\alpha}$ de la fila $\alpha$ como $\mu_{\alpha} = (s_{\alpha, 1}, s_{\alpha, 2}, \ldots, s_{\alpha, n})$, dando lugar a que una configuración del sistema sea una $n$-tupla $\sigma = (\mu_1, \mu_2, \ldots, \mu_n)$. Bajo este orden de ideas, el conjunto de todas las configuraciones posibles de una fila se define como $\Omega = \{(s_1, s_2, \ldots, s_n) | s_i \in \{-1, 1\}\} = \{-1, 1\}^n$, de modo que $\mu_{\alpha} \in \Omega$. Note que el conjunto de todas las configuraciones posibles del sistema es simplemente $\Omega^n$ = $\{-1, 1\}^{n^2}$.

Ahora, se imponen las condiciones de frontera periódicas

$$s_{\alpha, n+1} = s_{\alpha, 1} \quad \forall \alpha$$

y

$$\mu_{n+1} = \mu_1,$$

las cuales le otorgan al sistema una topología de toro. Concretamente, las condiciones de frontera se imparten agregando una fila y una columna al cuadrado y requiriendo que la configuración de la última fila sea igual a la de la primera fila y que la configuración de la última columna sea igual a la de la primera columna, i.e. las ecuaciones anteriores; la fila y la columna adicionales son únicamente una herramienta matemática para describir la topología del toro y no representan realmente sitios físicos en el sistema, por lo que en el Hamiltoniano se verá que solo se utilizan para representar las interacciones entre la última y la primera fila y entre la última y la primera columna.

Estrictamente hablando, las dos condiciones de frontera no son necesarias, realmente es posible derivar la formulación matricial del modelo construyendo simplemente un cilindro en lugar de un toro. Sin embargo, los resultados finales en ambos casos resultan ser los mismos y la simetría asociada a la forma de toro permite una solución menos complicada.

Definamos ahora la función de energía del sistema, i.e. el Hamiltoniano. Para esto, se definen primero dos funciones de energía, una para las interacciones entre espines adyacentes en la misma columna, i.e. para interacciones inter-fila, y otra para las interacciones entre espines adyacentes en la misma fila, i.e. para interacciones intra-fila, junto con el término de interacción de los espines de la fila con un campo magnético externo $h$. Acá, entonces, se introducen dos suposiciones adicionales: que tanto las interacciones inter-fila como las interacciones intra-fila son isotrópicas y que las interacciones entre espines son únicamente entre vecinos más cercanos. Dadas dichas suposiciones, la función de energía para las interacciones inter-fila es

$$E_{\text{inter}}(\mu_{\alpha}, \mu_{\alpha+1}) = - \epsilon \sum_{k=1}^{n} s_{\alpha, k} s_{\alpha+1, k}$$

y la función de energía para las interacciones intra-fila (junto con el término de campo magnético) es

$$E_{\text{intra}}(\mu_{\alpha}) = - \epsilon \sum_{k=1}^{n} s_{\alpha, k} s_{\alpha, k+1} - h \sum_{k=1}^{n} s_{\alpha, k}.$$

Note que $E_{\text{inter}}$ es simétrica en sus dos argumentos, por lo que $E_{\text{inter}}(\mu, \mu') = E_{\text{inter}}(\mu', \mu)$.

Luego, la energía total del sistema es

$$E_I(\mu_1, \ldots, \mu_n) = \sum_{\alpha=1}^{n} \Big[E_{\text{inter}}(\mu_{\alpha}, \mu_{\alpha+1}) + E_{\text{intra}}(\mu_{\alpha}) \Big],$$

donde claramente se ve el uso de las condiciones de frontera periódicas para representar las interacciones entre la última y la primera fila y entre la última y la primera columna.

La función de partición canónica del sistema, consecuentemente, es
\begin{align*}
    Q_I (h, T) &= \sum_{\mu_1 \in \Omega} \cdots \sum_{\mu_n \in \Omega} \exp \Big\{-\beta E_I(\mu_1, \ldots, \mu_n) \Big\}\\
    &= \sum_{\mu_1 \in \Omega} \cdots \sum_{\mu_n \in \Omega} \exp \Big\{-\beta \sum_{\alpha=1}^{n} \Big[E_{\text{inter}}(\mu_{\alpha}, \mu_{\alpha+1}) + E_{\text{intra}}(\mu_{\alpha}) \Big] \Big\}
\end{align*}

donde la suma resultante tiene $2^{n^2}$ términos, lo cual es consistente con el tamaño del conjunto de configuraciones del sistema $\Omega^n$.

## **2. Asociación de Espacios Vectoriales y Formulación Matricial**

### **2.1 Espacio Vectorial de Configuraciones**

Para poder realizar la formulación matricial del modelo de forma rigurosa, es necesario asociarle a la definición previa una estructura vectorial. En realidad, esta construcción es bastante natural e intuitiva, pero es importante hacerla de forma explícita para darle un sentido claro a la formulación matricial.

Para empezar, en cada sitio $x \in \Lambda$ se define un espacio de Hilbert $V$ de dimensión 2 sobre $\mathbb{C}$ o $\mathbb{R}$, realmente no importa porque los objetos relevantes para la formulación terminan siendo reales; el espacio es el mismo para todos los sitios. Luego, se escoge una base ortonormal $\{|+\rangle, |-\rangle\}$ para $V$, realizando la identificación $|+\rangle \leftrightarrow +1$ y $|-\rangle \leftrightarrow -1$.

Con base en dicho espacio, se puede definir un espacio de Hilbert $\mathcal{H}$ para cada fila $\alpha$ del sistema como el producto tensorial

$$\mathcal{H} = V^{\otimes n},$$

el cual es de dimensión $2^n$.

La base ortonormal canónica de $\mathcal{H}$ es

$$\mathcal{B} = \{ |\mu\rangle = |s_1\rangle \otimes \cdots \otimes |s_n\rangle | \mu \text{ está asociado a } \mu_{\alpha} = (s_1, \cdots, s_n) \in \Omega \text{ para la fila de interés } \alpha\},$$

por lo que hay una correspondencia biyectiva entre los vectores base de $\mathcal{H}$ y las configuraciones posibles de una fila del sistema.

### **2.2. Definición del Operador de Transferencia y Formulación Matricial de la Función de Partición**

La función de partición canónica del sistema es

$$Q_I (h, T) = \sum_{\mu_1 \in \Omega} \cdots \sum_{\mu_n \in \Omega} \exp \Big\{-\beta \sum_{\alpha=1}^{n} \Big[E_{\text{inter}}(\mu_{\alpha}, \mu_{\alpha+1}) + E_{\text{intra}}(\mu_{\alpha}) \Big] \Big\}.$$

Entonces, definiendo el operador lineal $P: \mathcal{H} \to \mathcal{H}$ como

$$\langle \mu | P | \mu' \rangle = \exp \Big\{-\beta \Big[E_{\text{inter}}(\mu, \mu') + E_{\text{intra}}(\mu) \Big] \Big\},$$

se tiene que

\begin{align*}
    Q_I (h, T) &= \sum_{\mu_1 \in \Omega} \cdots \sum_{\mu_n \in \Omega} \prod_{\alpha=1}^{n} \langle \mu_{\alpha} | P | \mu_{\alpha+1} \rangle\\
    &= \sum_{\mu_1, \mu_2, \cdots, \mu_n} \langle \mu_1 | P | \mu_2 \rangle \langle \mu_2 | P | \mu_3 \rangle \cdots \langle \mu_n | P | \mu_1 \rangle.
\end{align*}

Note que la representación matricial de $P$ en la base $\mathcal{B}$ es una matriz real de $2^n \times 2^n$.

Recuerde que la identidad se puede escribir en términos de productos tensoriales en la base $\{|\mu\rangle \langle \mu|\}$ como $I = \sum_{\mu} |\mu\rangle \langle \mu|$.

Considere, entonces, la traza

$$\text{Tr} \left(P^n \right) = \sum_{\mu_1} \langle \mu_1 | P^n | \mu_1 \rangle.$$

Introduciendo $n-1$ copias de la identidad entre cada par de operadores $P$ en la traza, se tiene que

\begin{align*}
    \text{Tr} \left(P^n \right) &= \sum_{\mu_1} \langle \mu_1 | P^n | \mu_1 \rangle\\
    &= \sum_{\mu_1} \langle \mu_1 | P I P I \cdots P I P I P | \mu_1 \rangle\\
    &= \sum_{\mu_1} \langle \mu_1 | P \left(\sum_{\mu_2} |\mu_2\rangle \langle \mu_2| \right) P \left(\sum_{\mu_3} |\mu_3\rangle \langle \mu_3| \right) \cdots P \left(\sum_{\mu_{n-1}} |\mu_{n-1}\rangle \langle \mu_{n-1}| \right) P \left(\sum_{\mu_n} |\mu_n\rangle \langle \mu_n| \right) P | \mu_1 \rangle\\
    &= \sum_{\mu_1} \langle \mu_1 | P \left(\sum_{\mu_2} |\mu_2\rangle \langle \mu_2| \right) P \left(\sum_{\mu_3} |\mu_3\rangle \langle \mu_3| \right) \cdots P \left(\sum_{\mu_{n-1}} |\mu_{n-1}\rangle \langle \mu_{n-1}| \right) P \left(\sum_{\mu_n} \langle \mu_n| P | \mu_1 \rangle |\mu_n\rangle \right)\\
    &= \sum_{\mu_1} \langle \mu_1 | P \left(\sum_{\mu_2} |\mu_2\rangle \langle \mu_2| \right) P \left(\sum_{\mu_3} |\mu_3\rangle \langle \mu_3| \right) \cdots P \left(\sum_{\mu_{n-1}} |\mu_{n-1}\rangle \langle \mu_{n-1}| \right) \left(\sum_{\mu_n} \langle \mu_n| P | \mu_1 \rangle P |\mu_n\rangle \right)\\
    &= \sum_{\mu_1} \langle \mu_1 | P \left(\sum_{\mu_2} |\mu_2\rangle \langle \mu_2| \right) P \left(\sum_{\mu_3} |\mu_3\rangle \langle \mu_3| \right) \cdots \left(\sum_{\mu_{n-1}} \sum_{\mu_n} \langle \mu_n| P | \mu_1 \rangle \langle \mu_{n-1}| P | \mu_n \rangle P |\mu_{n-1}\rangle \right)\\
    &= \sum_{\mu_1} \langle \mu_1 | P \left(\sum_{\mu_2} |\mu_2\rangle \langle \mu_2| \right) \left(\sum_{\mu_3} \langle \mu_3| P | \mu_4 \rangle P |\mu_3\rangle \right) \cdots \left(\sum_{\mu_{n-1}} \sum_{\mu_n} \langle \mu_{n-1}| P | \mu_n \rangle \langle \mu_n| P | \mu_1 \rangle \right)\\
    &= \sum_{\mu_1} \langle \mu_1 | \left(\sum_{\mu_2} \sum_{\mu_3} \langle \mu_2| P | \mu_3 \rangle \langle \mu_3| P | \mu_4 \rangle P |\mu_2\rangle \right) \cdots \left(\sum_{\mu_{n-1}} \sum_{\mu_n} \langle \mu_{n-1}| P | \mu_n \rangle \langle \mu_n| P | \mu_1 \rangle \right)\\
    &= \left(\sum_{\mu_1} \sum_{\mu_2} \sum_{\mu_3} \langle \mu_1 | P | \mu_2 \rangle \langle \mu_2 | P | \mu_3 \rangle \langle \mu_3 | P | \mu_4 \rangle \right) \cdots \left(\sum_{\mu_{n-1}} \sum_{\mu_n} \langle \mu_{n-1}| P | \mu_n \rangle \langle \mu_n| P | \mu_1 \rangle \right)\\
    &= \sum_{\mu_1, \mu_2, \ldots, \mu_n} \langle \mu_1 | P | \mu_2 \rangle \langle \mu_2 | P | \mu_3 \rangle \cdots \langle \mu_n | P | \mu_1 \rangle\\
    &= Q_I (h, T).
\end{align*}

Por lo tanto,

$$Q_I (h, T) = \text{Tr} \left(P^n \right) = \sum_{\alpha=1}^{2^n} \left(\lambda_{\alpha} \right)^n,$$

de modo que el problema de reduce a encontrar los valores propios de $P$, i.e un problema de valores propios en un espacio de dimensión $2^n$.

Este resultado aplica en general para $h \neq 0$.

### **2.3 Interpretación del Operador de Transferencia**

A partir de la definición de $P$, se puede ver que el operador actúa de la siguiente forma sobre un vector base de $\mathcal{H}$:

$$P |\mu'\rangle = \sum_{\mu} \langle \mu | P | \mu' \rangle |\mu\rangle = \sum_{\mu} \exp \Big\{-\beta \Big[E_{\text{inter}}(\mu, \mu') + E_{\text{intra}}(\mu) \Big] \Big\} |\mu\rangle.$$

Lo anterior es sencillo de visualizar en la representación de coordenadas asociada a la base $\mathcal{B}$.

En palabras, la acción del operador de transferencia sobre un estado de una fila del sistema resulta en una combinación lineal real de todos los estados posibles de una fila del sistema. La interpretación física de lo anterior es clara si se considera que, dada una fila particular $\alpha$, el operador de transferencia actúa sobre el estado $\mu'$ de la fila $\alpha$ y genera un vector que vive en la copia de $\mathcal{H}$ asociada a la fila $\alpha+1$. De esta forma, el operador de transferencia "transfiere" la información de la fila $\alpha$ a la fila $\alpha+1$. Si ce cierra $P | \mu' \rangle$ con un bra $\langle \mu''|$ arbitrario, por ejemplo, se observa que los estados $\langle \mu''|$ de la fila $\alpha+1$ dependen del estado $\mu'$ de la fila $\alpha$ y de la configuración de espines que tendría la fila $\alpha+1$ si se fijara $\langle \mu''|$.

Ahora, para interpretar los coeficientes de la combinación lineal, se puede pensar el término

$$\langle \mu | P | \mu' \rangle = \exp \Big\{-\beta \Big[E_{\text{inter}}(\mu, \mu') + E_{\text{intra}}(\mu') \Big] \Big\}$$


como un factor asociado a la probabilidad de que la fila $\alpha+1$ esté en el estado $\mu$ dado que la fila $\alpha$ está en el estado $\mu'$. Para justificar lo anterior, considere la probabilidad de que el sistema se encuentre en una configuración particular $\sigma = (\mu_1, \mu_2, \ldots, \mu_n)$, la cual se puede escribir como

$$\mathcal{P}(\sigma) = \frac{1}{\text{Tr}(P^n)} \langle \mu_1 | P | \mu_2 \rangle \langle \mu_2 | P | \mu_3 \rangle \cdots \langle \mu_n | P | \mu_1 \rangle.$$

Lo que sigue es netamente un "heuristic" argument, i.e. un argumento pedagógico-intuitivo que no es estrictamente fiel a la "realidad" física del modelo.

Si ahora consideramos una construcción del toro, más matemática que física, en la que secuencialmente se van agregando filas al sistema, entonces tiene sentido hablar de la probabilidad de que una fila particular esté en un estado $\mu'$ dado que la fila anterior está en un estado $\mu$. Para darle un sabor más físico a la construcción podemos pensar que cogemos una fila inicial y dejamos que entre en equilibrio, luego, espontáneamente, agregamos una fila adicional y dejamos que entre en equilibrio con la fila anterior, y así sucesivamente hasta que se agregan todas las filas; incluso si esto le molesta, por ejemplo porque estamos agregando interacciones que deberían afectar a la fila anterior, se puede imaginar que la construcción se hace más bien agregando espines individuales a cada fila, empezando a crear una nueva fila cuando se han agregado n espines, y que al ser agregado un espín, si la cadena es muy grande (como debería ser en un sistema termodinámico), este no afecta al resto de la fila, sin embargo, estas dos construcciones tienen problemas conceptuales si uno lo piensa con cuidado, razón por la cual se hizo énfasis en que es una construcción pedagógica. En este sentido, la probabilidad de que la fila $\alpha+1$ esté en un estado $\mu'$ dado que la fila $\alpha$ está en un estado $\mu$ es proporcional al factor

$$\exp \Big\{-\beta \Big[E_{\text{inter}}(\mu, \mu') + E_{\text{intra}}(\mu') \Big] \Big\} = \langle \mu' | P | \mu \rangle,$$

por lo que se ve que los elementos matriciales de $P$ están íntimamente relacionados con la propagación de información a lo largo del sistema mediante unas probabilidades condicionales. Ahora, estrictamente hablando, no tiene tanto sentido pensar acerca de construcciones secuenciales debido a que se tendrían que fijar, de alguna forma, ciertos estados de equilibrio en las filas antecedentes a la nueva fila, pero para asociar la intuición desarrollada sobre $P$ con una interpretación más "realista", basta con identificar que $P$ captura la propagación de información estadística a lo largo del sistema teniendo en cuenta todas las configuraciones posibles del sistema (que, al tener en cuenta la fila y columna $n+1$ son cíclicas, ergo la topología de toro), i.e. todas las combinaciones posibles de estados de equilibrio de las filas. De esta forma, no hay una propagación "real" entre las filas, sino más bien una propagación abstracta y matemática de información probabilística asociada a los estados posibles de equilibrio; dicha interpretación más "realista" se encuentra al observar que, a partir de la intuición ganada con nuestra construcción pedagógica, en la expresión para la probabilidad de una configuración de todo el sistema se observa que la traza de $P^n$ está teniendo en cuenta todas esas posibles  probabilidades condicionales al variar matemáticamente los estados de las filas previas. De hecho, de forma puramente matemática, si hay una noción de recorrer el toro probando diferentes configuraciones de filas y propagando información estadística con las probabilidades condicionales mediante el operador de transferencia, esta se puede extraer de sus elementos matriciales y viendo el toro:

Para esto, identifiquemos por convención, sin pérdida de generalidad, que la fila 1 es el circulo interno de la figura del toro y que la fila $n$ es el círculo externo. Entonces, el elemento matricial $\langle \mu_{2} | P | \mu_{1} \rangle$ es

$$\exp \Big\{-\beta \Big[E_{\text{inter}}(\mu_{1}, \mu_{2}) + E_{\text{intra}}(\mu_{2}) \Big] \Big\},$$

por lo que hay una propagación de información de la fila 1 a la fila 2 vía $E_{\text{inter}}$ y además, en términos de probabilidades condicionales, una propagación de información probabilística de la fila 1 a la fila 2. Además, como

$$E_{\text{inter}} (\mu_{1}, \mu_{2}) = - \epsilon \sum_{k=1}^{n} s_{1, k} s_{2, k},$$

se observa que gráficamente podemos imaginarnos que estamos recorriendo los primeros dos círculos (primeros desde adentro hacia afuera del toro) del toro y que estamos viendo las interacciones entre los espines de la fila 1 con los de la fila 2 y, mediante $E_{\text{intra}}$, las interacciones entre los espines que conforman la fila 2. Luego, al terminar de recorrer dichos círculos, se pasa mediante $\langle \mu_{3} | P | \mu_{2} \rangle$ a recorrer el segundo y tercer circulo interno y así sucesivamente hasta que se llega al círculo exterior correspondiente a la fila $n$, recorrido en el cual se tienen en cuenta las interacciones entre los espines que conforman la fila $1$  porque

$$\langle \mu_{n+1} | P | \mu_{n} \rangle = \langle \mu_{1} | P | \mu_{n} \rangle = \exp \Big\{-\beta \Big[E_{\text{inter}}(\mu_{n}, \mu_{1}) + E_{\text{intra}}(\mu_{1}) \Big] \Big\}.$$

En este sentido se observa que uno si recorre el toro, pero de forma matemática en vez de física, y que en los recorridos se prueban todas las configuraciones posibles del sistema, lo cual sucede mediante aplicaciones sucesivas de $P$ sobre los estados de la primera fila. Lo anterior termina de ser claro una vez uno recuerda que las sumatorias en la expresión

$$Q_I (h, T) = \sum_{\mu_1, \mu_2, \ldots, \mu_n} \langle \mu_1 | P | \mu_2 \rangle \langle \mu_2 | P | \mu_3 \rangle \cdots \langle \mu_n | P | \mu_1 \rangle$$

iteran sobre todos los estados posibles de las filas, mostrando que $P$ está propagando información estadística a lo largo del sistema para todas las configuraciones posibles del sistema. Finalmente, en la expresión

$$Q_I (h, T) = \text{Tr} \left(P^n \right) = \sum_{\mu_1} \langle \mu_1 | P^n | \mu_1 \rangle,$$

se ve que eso es equivalente a lo mencionado de considerar aplicaciones sucesivas del operador de transferencia sobre los estados de la primera fila; en este caso la elección de una "primera" fila es puramente una cuestión de convención po la naturaleza cíclica del toro, pero su armara por ejemplo un cilindro y las filas son secciones transversales del cilindro, entonces la elección de la primera fila es una elección real en lo que concierne a la estructura geométrica del sistema.

De hecho, en esa simetría traslacional, la cual es igualmente clara en las definiciones de $E_{\text{inter}}$ y $E_{\text{intra}}$, obtenemos pistas sobre por qué es menos complicado resolver este problema que el de un sistema con (algunas) fronteras abiertas.

**Nota:** Uno podría, otra vez con propósitos pedagógicos, complementar la interpretación de los recorridos del toro mediante aplicaciones de $P$, asociadas en cierto sentido a probabilidades condicionales, haciendo una analogía con cadenas de Markov.

# **Álgebras de Clifford y Espinores**

Sea $V$ un espacio vectorial de dimensión $n$ sobre $\mathbb{R}$ dotado de una forma bilineal simétrica $g: V \times V \to \mathbb{R}$, $A$ un álgebra asociativa con identidad $1_A$ y $\gamma: V \to A$ un mapa lineal.

**Definición**

El par ($A$, $\gamma$) es un álgebra de Clifford para el espacio cuadrático
($V$, $g$) si se satisfacen las siguientes condiciones:

1. $A$ es generado como álgebra por $\{\gamma(v)|v \in V\}$ y $\{a1_A|a \in \mathbb{R}\}$.
1. $\gamma(v)\gamma(u) + \gamma(u)\gamma(v) = 2g(v, u)1_A$ para todo $v, u \in V$.

Las álgebras de Clifford también son conocidas como álgebras geométricas, lo cual cobra sentido al observar que la relación entre los $\gamma$'s involucra la forma bilineal simétrica $g$, por lo que es evidente que las álgebras de Clifford buscan codificar la información geométrica del espacio cuadrático al tener en cuenta la estructura adicional que aporta $g$ a $V$.

Se resalta que la condición sobre el mapa $\gamma$ es equivalente a la condición de que $\gamma(v)^2 = Q(v)1_A$ se satisfaga para todo $v \in V$, donde $Q(v) = g(v, v)$ es la forma cuadrática inducida por $g$; esta forma de la condición $2$ también es comúnmente utilizada.

Informalmente, se puede pensar acerca de $\gamma$ como una especie de "raíz cuadrada" de $Q$. Es común referirse a $\gamma$ como un"Clifford mapping".

Si ahora se considera una base ortonormal $\mathfrak{B} = \{e_1, e_2, \ldots, e_n\}$ de $V$, entonces en el álgebra de Clifford $(A, \gamma)$ se tiene que

$$\gamma(e_i)\gamma(e_j) + \gamma(e_j)\gamma(e_i) = 0_{A}, \quad i \neq j$$

y

$$\gamma(e_i)^2 = Q(e_i)1_A = 1_A.$$

A partir de dichas relaciones, cualquier producto involucrado a los $\gamma(e_i)$ (con $i = 1, 2, \ldots, n$) y sus potencias se puede reordenar de la forma

$$\gamma(e_1)^{\mu_1}\gamma(e_2)^{\mu_2}\cdots\gamma(e_n)^{\mu_n}, \quad \mu_i = 0, 1 \quad (i = 1, 2, \ldots, n),$$

donde $\gamma(e_1)^0\gamma(e_2)^0\cdots\gamma(e_n)^0 = 1_A$. Entonces, como por definición $\{\gamma(v)|v \in V\}$ y $\{a1_A|a \in \mathbb{R}\}$ generan como álgebra a $A$, se tiene que esta es generada por los productos anteriores, i.e.

$$A = \text{span}\{\gamma(e_1)^{\mu_1}\gamma(e_2)^{\mu_2}\cdots\gamma(e_n)^{\mu_n} | \mu_i = 0, 1\}.$$

Consecuentemente, dado que la cantidad de combinaciones posibles de productos de los $\gamma(e_i)$ es $2^n$, se tiene que $\text{dim}(A) \leq 2^n$ (la dimensión de $A$ hace referencia a la dimensión del espacio vectorial subyacente) porque, aunque es cierto que el conjunto de productos de los $\gamma(e_i)$ genera a $A$ como espacio vectorial (es un conjunto generador del espacio vectorial subyacente), sus elementos pueden ser linealmente dependientes, por lo que el conjunto no es necesariamente una base de $A$.

Las álgebras de Clifford con dimensión máxima tienen una propiedad especial llamada "universalidad", la cual las distingue de las álgebras de Clifford con dimensión menor. Dichas álgebras de Clifford son únicas salvo por un isomorfismo de álgebras, por lo que se habla de "el" álgebra de Clifford universal asociada a un espacio cuadrático.

Existen teoremas que permiten asociarle a todo espacio cuadrático un álgebra de Clifford universal.

Dado un espacio vectorial real $V$ de dimensión $m$, su complexificación $V_{\mathbb{C}} = \mathbb{C} \otimes V$ es un espacio vectorial complejo de dimensión real $2m$ y dimensión compleja $m$.

Hubiera sido posible definir inicialmente las álgebras de Clifford sobre cualquier campo, pero gracias a teoremas de isomorfismos, basta con definir las álgebras de Clifford sobre $\mathbb{R}$. Un ejemplo relevante de esto se ve al tomar un espacio cuadrático complejo $(V_{\mathbb{C}}, g_{\mathbb{C}})$ y su álgebra de Clifford universal $Cl(V_{\mathbb{C}}, g_{\mathbb{C}})$. Resulta que esta álgebra de Clifford es isomorfa a la complexificación del álgebra de Clifford universal asociada al espacio cuadrático real $(V, g)$, i.e.

$$\mathbb{C} \otimes Cl(V,g) = Cl_\mathbb{C}(V,g) \cong Cl_{\mathbb{C}}(V, g).$$

Dado que a cualquier espacio cuadrático se le puede asociar un álgebra universal de Clifford, dada una forma bilineal simétrica $g$ en $R^n$ con signatura $(p, q)$, se puede hablar del álgebra universal de Clifford $Cl_{p, q} = Cl(\mathbb{R}^{p, q})$.

Entonces, dado $2n = p + q$, i.e. la dimensión es par, es un resultado estándar que la complexificación de $Cl_{p, q}$ es algebraicamente isomorfa al álgebra de matrices complejas $M(2^n,\mathbb{C})$, i.e.

$$Cl_{\mathbb{C}}(2n) \cong M(2^n, \mathbb{C}).$$

Luego, si ahora escogemos un espacio euclidiano de dimensión par $\mathbb{R}^{2n}$ con su producto interno usual, se tiene que, mediante el uso de complexificación, se le puede asociar, en cierto sentido el álgebra de Clifford universal $Cl_{\mathbb{C}}(2n)$ y, por ende, $M(2^n, \mathbb{C})$; esta álgebra claramente está asociada a $\mathbb{C}^{2^n}$ porque la complexificación de $R^m$ es isomorfa a $\mathbb{C}^m$.

Finalmente, se menciona que la asociación a $\mathbb{C}^{2^n}$ tiene una interpretación profunda puesto que resulta que el espacio de espinores algebraicos asociado a $Cl_{\mathbb{C}}(2n)$, sobre los cuales no se va a profundizar, es isomorfo a $\mathbb{C}^{2^n}$. Es decir que las matrices complejas de $2^n \times 2^n$ actúan sobre los espinores algebraicos de $Cl_{\mathbb{C}}(2n)$.

# **Las Matrices de Weyl-Brauer o Matrices $\Gamma$**

De la expresión expandida de los elementos matriciales de $P$ tenemos:

$$\langle s_1, s_2, \ldots, s_n | P | s_1', s_2', \ldots, s_n' \rangle = \prod_{k=1}^{n} e^{\beta h s_k} e^{\beta \epsilon s_k s_{k+1}} e^{\beta \epsilon s_k s_k'}.$$

Definiendo $V_1'$, $V_2$ y $V_3$ como

$$\langle s_1, s_2, \ldots, s_n | V_1' | s_1', s_2', \ldots, s_n' \rangle = \prod_{k=1}^{n} e^{\beta \epsilon s_k s_k'},$$

$$\langle s_1, s_2, \ldots, s_n | V_2 | s_1', s_2', \ldots, s_n' \rangle = \delta_{s_1, s_1'} \ldots \delta_{s_n, s_n'} \prod_{k=1}^{n} e^{\beta \epsilon s_k s_{k+1}},$$

y

$$\langle s_1, s_2, \ldots, s_n | V_3 | s_1', s_2', \ldots, s_n' \rangle = \delta_{s_1, s_1'} \ldots \delta_{s_n, s_n'} \prod_{k=1}^{n} e^{\beta h s_k}$$

es evidente que

$$P = V_3 V_2 V_1',$$

esto es

$$\langle \mu | P | \mu' \rangle = \sum_{\mu''} \sum_{\mu'''} \langle \mu | V_3 | \mu'' \rangle \langle \mu'' | V_2 | \mu''' \rangle \langle \mu''' | V_1' | \mu' \rangle.$$

Bajo ciertas manipulaciones adicionales, no triviales, usando las matrices de Pauli es posible deducir que

$$P = [2\sinh(2\beta \epsilon)]^{n/2} V_3 V_2 V_1$$

para una matriz $V_1$ que depende de las matrices de Pauli.

Cuando $h=0$, se tiene que $V_3 = 1$, por lo que

$$P = [2\sinh(2\beta \epsilon)]^{n/2} V_2 V_1.$$

Lo que motiva el estudio matemático de las matrices $\Gamma$ es que estas son una generalización de las matrices de Pauli y están íntimamente relacionadas con las álgebras de Clifford y los grupos de espín y de rotación como consecuencia, permitiendo utilizar las propiedades algebraicas y de teoría de grupos de las álgebras de Clifford para diagonalizar $P$.

HISTORIA DE PAPER DE WEYL-BRAUER

Existen $2n$ matrices $\Gamma_{\mu}$ ($\mu = 1, 2, \ldots, 2n$) que satisfacen las relaciones de anticonmutación de las álgebras de Clifford

$$\Gamma_{\mu}\Gamma_{\nu} + \Gamma_{\nu}\Gamma_{\mu} = 2\delta_{\mu\nu}$$

El paper demostró las siguientes propiedades de las matrices $\Gamma$, las cuales son parcialmente intuitivas al considerar lo que se mencionó sobre álgebras de Clifford:

(a) La dimensionalidad de $\Gamma_{\mu}$ no puede ser menor que $2^n$.

(b) Si {$\Gamma_{\mu}$} y {$\Gamma_{\mu}'$} son dos conjuntos de matrices que satisfacen las relaciones de anticonmutación, entonces existe una matriz no singular $S$ tal que $\Gamma_{\mu} = S\Gamma_{\mu}'S^{-1}$ y viceversa.

(c) Cualquier matriz $2^n \times 2^n$ es una combinación lineal de la identidad, las matrices $\Gamma_{\mu}$ en su versión $2^n \times 2^n$ y los productos independientes $\Gamma_{\mu}\Gamma_{\nu}, \Gamma_{\mu}\Gamma_{\nu}\Gamma_{\lambda}, \ldots$.

Las matrices de Pauli son el caso especial de estas matrices en el que $n=1$.

Las matrices $\gamma_{\mu}$ de Dirac son el caso $n=4$, las cuales tienen aplicaciones en el espaciotiempo de Minkowski, en particular, en la ecuación de onda relativista de Dirac.

Lo que sigue se relaciona con que $V_1$ y $V_2$ resultaran ser matrices que transforman $\{\Gamma_{\mu}\}$ en conjuntos equivalentes.

Si se aplica una matriz $2n \times 2n$ ortogonal $\omega$ a las matrices de Weyl-Brauer, se obtiene un conjunto de matrices equivalentes a las originales, i.e.

$$\Gamma_{\mu}' = \sum_{\nu=1}^{2n} \omega_{\mu\nu}\Gamma_{\nu}.$$

Resulta que $\Gamma_{\mu}'$ también satisface las relaciones de anticonmutación por la ortogonalidad de $\omega$.

Por lo tanto, por la propiedad (b) se tiene que existe una matriz no singular $S$ tal que

$$\Gamma_{\mu}' = S(\omega)\Gamma_{\mu}S^{-1}(\omega)$$

para $S(\omega)$ una matriz $2^n \times 2^n$.

Entonces, via las matrices $\Gamma$ hay una identificación entre

$$\omega \leftrightarrow S(\omega).$$
